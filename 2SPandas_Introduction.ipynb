Aggregation Function in Pandas Aggreation- Computing a summary statistic about each group


```python
dir(df) # function that can be applied on a dataframe
```

Common Aggreation Functions are: min,max,count,sum,prod,mean,median,mode,mad,std,var

```python
# Compute group sums or means
# Compute group sizes or counts
# agg() or aggreate()- useful when multiple statistics are computed per column
# Here's how to calculate the Mean Absoute Devaition - mad()
#Step 1:Calculate the mean
#Step 2:Calculate how far away each data point is from the mean using positive distances. These are called as absolute deviation
#Step 3:Add those devaitions together.
#Step 4: Divide the sum by the number of data points.

df[['phd','salary']].agg(['min','mean','max','mad','std'])


```

```python
df[['salary']].agg('skew')
```

```python
df[['salary']].agg('kurt')
```
